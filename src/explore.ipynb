```python
# Your code here
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

url = "https://raw.githubusercontent.com/4GeeksAcademy/naive-bayes-project-tutorial/main/playstore_reviews.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop("package_name", axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df["review"] = df["review"].str.strip().str.lower()
```

```python
df.head()
```

```python

#Convertimos en train y test
ex = df["review"]
ey = df["polarity"]

X_train, X_test, y_train, y_test = train_test_split(ex, ey, test_size = 0.2, random_state = 42)

X_train.head()
```

```python
vec_model = CountVectorizer(stop_words = "english")
X_train = vec_model.fit_transform(X_train).toarray()
X_test = vec_model.transform(X_test).toarray()

X_train
```

```python
#Empiezo con el modelo

from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB

model = MultinomialNB()
model2 = GaussianNB()
model3 = BernoulliNB()
model.fit(X_train, y_train)
```

```python
model2.fit(X_train, y_train)
```

```python
model3.fit(X_train, y_train)
```

```python
#Multi
y_pred = model.predict(X_test)
y_pred
```

```python
#Gaussian
y_pred2 = model2.predict(X_test)
y_pred2
```

```python
#Berno
y_pred3 = model3.predict(X_test)
y_pred3
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred2)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred3)
```

```python
#He probado con los tres modelos y se puede ver claramente que el mejor es el multinomial.
#Esto se podía preveer ya que el multinomial cuando se trata de texto y buscar palabras funciona mejor.
#El gaussian es para datos continuos y el bernoulli es para datos binarios.
#Ahora vamos a mejorar el modelo

from sklearn.model_selection import RandomizedSearchCV

parametros = {
    "alpha": np.linspace(0.01, 10.0, 200),
    "fit_prior": [True, False]
}

buscador = RandomizedSearchCV(model, parametros, n_iter = 50, scoring = "accuracy", cv = 5, random_state = 42)
buscador
```

```python
#Usamos el mejor anterior obviamente

buscador.fit(X_train, y_train)

print(f"Los mejores parametros: {buscador.best_params_}")
```

```python
#Aplicamos los parametros anteriores

model = MultinomialNB(alpha = 1.917638190954774, fit_prior = False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)

#Se puede ver una pequeña mejora poniendo los hiperparametros.
```

```python
from pickle import dump

dump(model, open("../models/bayes.sav", "wb"))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Precisión:", accuracy_score(y_test, y_pred))

#Con regresión lineal podemos mejorar un poco más el modelo y pasar de 0.82 a 0.83 

dump(model, open("../models/lineal.sav", "wb"))
```
